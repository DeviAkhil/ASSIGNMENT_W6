## 2. Find out how many countries with respect to the region are available in the  dataset

```python
data['Region'].value_counts()
```

```python
region=data['Region'].value_counts()
region=dict(region)
reg=list(region.keys())
values=list(region.values())
```

```python
plt.figure(figsize=(15,15))
plt.bar(reg,values,color ='maroon',width = 0.8)
plt.xlabel("Region")
plt.xticks(rotation='vertical')
plt.yticks(np.arange(0,20,2))
plt.ylabel("No:of countries")
plt.show()
```

## 3. On average which region has the highest

### a. GDP per capita (current US$)

```python
gdp=data.groupby(by='Region')['GDP per capita (current US$)'].mean().sort_values()
gdp=dict(gdp)
region=list(gdp.keys())
values_1=list(gdp.values())
```

```python
plt.figure(figsize=(15,15))
plt.bar(region,values_1,color ='b',width = 0.8)
plt.xlabel("Region")
plt.xticks(rotation='vertical')
plt.ylabel("GDP Average")
plt.show()
```

### Northern America has highest GDP per capita

### b. International trade: Imports (million US$)

```python
trades=data.groupby(by='Region')['International trade: Imports (million US$)'].mean().sort_values()
trades=dict(trades)
region_1=list(trades.keys())
values_11=list(trades.values())
```

```python
plt.figure(figsize=(10,10))
plt.bar(region_1,values_11,color ='g',width = 0.8)
plt.xlabel("Region")
plt.xticks(rotation='vertical')
plt.ylabel("International trade: Imports (million US$)")
plt.show()
```

### Nortern america has highest international trades

## 4. Find out which region has more consistent with respect to GDP per capita (current US$)

```python
gdp_std=data.groupby(by='Region')['GDP per capita (current US$)'].std().sort_values()
gdp_std=dict(gdp_std)
region_std=list(gdp_std.keys())
values_std=list(gdp_std.values())
```

```python
plt.figure(figsize=(10,10))
plt.bar(region_std,values_std,color ='r',width = 0.8)
plt.xlabel("Region")
plt.xticks(rotation='vertical')
plt.ylabel("International trade: Imports (million US$)")
plt.show()
```

### As per the graph Western africa has less deviation in GDP per capita
